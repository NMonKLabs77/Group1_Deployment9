### Amount of revenue generated for each order item (considering all order is paid)

```python

%%sql
---q2.2
SELECT ordered_item as 'Ordered Item', count(ordered_item) as 'Count',
SUM(total_price) as 'Total Revenue (INR)', (SUM(total_price) *0.012016) as 'Total Revenue (USD)'  
FROM account_ordermodel
GROUP BY ordered_item
;
```

### checking if name in billing address is similar to account_stripemodel table

```python
%%sql

SELECT account_billingaddress.name, account_billingaddress.state, account_stripemodel.name_on_card, account_stripemodel.address_state
FROM account_billingaddress
INNER JOIN account_stripemodel
ON account_billingaddress.name = account_stripemodel.name_on_card
LIMIT 10;
```

### Retriving the amount of order and total revenue per state (cosidered all order are paid)

```python
%%sql
-- revenue per state
-- q3
SELECT state, count(*) as 'Number of Products Ordered', SUM(total_price) as 'Total Revenue (INR)',
(SUM(total_price) *0.012016) as 'Total Revenue (USD)' 
FROM account_ordermodel
LEFT JOIN account_billingaddress
ON account_ordermodel.name = account_billingaddress.name
GROUP BY state;
```

### seeing if account_ordermodel state would match up with account_billing address state


```python
%%sql 
SELECT address, state
FROM account_ordermodel
LEFT JOIN account_billingaddress
ON account_ordermodel.name = account_billingaddress.name
WHERE account_ordermodel.name = 'Glenn Brown';
```

### Creating for a view of how product rank for each state

```python
%%sql
CREATE VIEW product_rank as 
SELECT state, ordered_item,count(*), rank() OVER (PARTITION BY state ORDER by count(*) desc) as rank
FROM account_ordermodel
LEFT JOIN account_billingaddress
ON account_ordermodel.name = account_billingaddress.name
GROUP BY state, ordered_item;
```

### Creating a table to show the top 5 rank items for each state

```python
%%sql 
SELECT * 
FROM product_rank
WHERE rank < 6;
```

### Order that was shipped out but not paid for

```python
%%sql
SELECT count(*),
SUM(total_price) as 'Total Revenue (INR)', (SUM(total_price) *0.012016) as 'Total Revenue (USD)' 
FROM account_ordermodel
WHERE paid_status = 0 AND is_delivered = 1
;
```
